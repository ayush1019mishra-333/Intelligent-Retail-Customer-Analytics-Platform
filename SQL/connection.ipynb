```python
import pandas as pd
import mysql.connector
from mysql.connector import Error

# Read CSV
df = pd.read_csv(r"C:\Users\amit6\OneDrive\Desktop\Intelligent-Retail-Customer-Analytics\Data\SQL_ready_data\transactions.csv")

# Connect to MySQL
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Amit@1019", 
    database="retail_customer_analytics"
)

cursor = conn.cursor()

query = """
INSERT INTO transactions
(TransactionID, InvoiceNo, CustomerID, StockCode, Quantity, UnitPrice, TotalAmount)
VALUES (%s,%s,%s,%s,%s,%s,%s)
"""

batch_size = 5000

for i in range(0, len(df), batch_size):
    batch = df.iloc[i:i+batch_size]

    data = list(batch.itertuples(index=False, name=None))

    cursor.executemany(query, data)
    conn.commit()

    print(f"Imported {min(i+batch_size, len(df))} / {len(df)} rows")

cursor.close()
conn.close()

print("Import Completed Successfully!")
```
