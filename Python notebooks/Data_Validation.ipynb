```python
last_date = df_1["InvoiceDate"].max()

df_2_modified = df_2[df_2["InvoiceDate"] > last_date]
df_2_modified
```

```python
retail_df = pd.concat([df_1,df_2_modified],ignore_index=True)
retail_df
```

```python
# retail_df = retail_df.drop_duplicates()
print("Shape before removing duplicates:", retail_df.shape)

duplicates = retail_df.duplicated().sum()
print("Duplicate rows:", duplicates)

retail_df = retail_df.drop_duplicates()

print("Shape after removing duplicates:", retail_df.shape)
```

### Duplicate Record Validation

After resolving the overlap between the two worksheets, the merged dataset was checked for exact duplicate rows.

- Total records before duplicate removal: **1,044,848**
- Exact duplicate rows identified: **11,812**
- Total records after duplicate removal: **1,033,036**

These duplicate records were removed to prevent double-counting of transactions in subsequent customer, sales, and forecasting analyses.
